# Iterative statements (Loops)

    Iterative statements, also known as loops, are used to execute a block of code repeatedly until a certain condition is met. There are two main types of iterative statements in Python: for loops and while loops.
    In this notebook we see how we can use while loop to avoid code redundancy

# While

    A while loop in Python is used to execute a block of code repeatedly as long as a given condition is true. The loop will continue running until the condition becomes false. It is essential to ensure that the condition can be satisfied at some point during the loop's execution; otherwise, it can lead to an infinite loop.

````python
# If we want to execute below block of code three times
print("welcome to my channel")
print('like')
print('Subscribe')
````

````python
# Without using loops
# This is considered as code redundancy: writing same code more no.of times
print("welcome to my channel")
print('like')
print('Subscribe')

print("welcome to my channel")
print('like')
print('Subscribe')

print("welcome to my channel")
print('like')
print('Subscribe')
````

````python
# By using loops we can actually print the result like below

initial_value = 1
while initial_value <= 3:
    print("welcome to my channel")
    print('like')
    print('Subscribe')
    initial_value += 1
````

````python
# If we don't mention a statement that effect the condition becomes False
# The code enters into infinite loop
initial_value = 1
while initial_value <= 3:
    print(initial_value)
````

    Python script to calculate the sum of all natural numbers from 0 up to a given positive integer entered by the user.

````python
initial_value = 0
addition = 0
num = int(input("enter the number: "))
while initial_value <= num:
    addition += initial_value
    initial_value += 1
print(addition)
````

````python
initial_value = 0
addition = 0
num = int(input("enter the number: "))
while initial_value <= num:
    addition += initial_value
    print('Adddition is ', addition)
    initial_value += 1
    print('Initial value is', initial_value)
print(addition)
````
