# output formatting

    Output formatting refers to the process of customizing the appearance and structure of output generated by a program. It is particularly essential when displaying information to users in a readable and organized manner.

```python
# considering some variables
name = 'Mahesh'
age = 21
location = 'India'
```

    I want to print the output of above variables like this meaningful information
    
    Name is Mahesh, age is 21 and location is India

## By placing variables wherever needed by sepeating text by ,(comma)

```python
print('Name is',name,', age is ',age, ' and location is ',location)
```

## using % approach

```python
print('Name is %s, age is %d and location is %s' %(name, age, location))
```

## using str.format()

```python
print("Name is {}, age is {} and location is {}".format(name,age, location))
```

    If we mistakenly placed variables, by proper indexing we can get the meaningful output

```python
print("Name is {1}, age is {0} and location is {2}".format(age,name, location)) 
# here age is given first then name but by mentioning indexed we can arrange them properly
```

## f-string

```python
print(f'Name is {name}, age is {age}, and location is {location}') # we can siply place variables wherever needed
```

## formatting tip

    In str.format() in str by using {:.2f} this way we can customize the values after the decimal, here I want 2 numbers after decimal

```python
value = 3.142352332
print('value is {:.2f}'.format(value))
```

```python
# if I have 
value1 = 3.14
print('value is {:.5f}'.format(value1)) # I want 5 values after the decimal --> so replaced with 0's
```
